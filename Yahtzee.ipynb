```python
import random

class Die:
    def __init__(self):
        self.sides = 6
        self.faces = list(range(1, self.sides+1))
        self._frozen = False
        self.roll()

    def roll(self):
        if not self._frozen:
            self.value = random.choice(self.faces)

    def freeze(self):
        self._frozen = True

    def __repr__(self):
        return str(self.value)

class _Category:
    def __repr__(self) -> str:
        raise NotImplementedError("Subclass must implement __repr__")

    def _satisfies(self, dice: list[Die]) -> bool:
        raise NotImplementedError("Subclass must implement _satisfies")

    def score(self, dice: list[Die]) -> int:
        raise NotImplementedError("Subclass must implement score")

class UpperCategory(_Category):
    NAMES = ['', 'Ones', 'Twos', 'Threes',
             'Fours', 'Fives', 'Sixes']

    def __init__(self, value):
        self.value = value

    def __repr__(self):
        return self.NAMES[self.value]

    def _satisfies(self, dice: list[Die]) -> bool:
        return sum(die.value for die in dice if die.value == self.value)

    def score(self, dice: list[Die]) -> int:
        return sum(die.value for die in dice if die.value == self.value)

class Chance(_Category):
    def __repr__(self):
        return "Chance"

    def _satisfies(self, dice: list[Die]) -> bool:
        return True

    def score(self, dice: list[Die]) -> int:
        return sum(die.value for die in dice)

class FullHouse(_Category):
    def __repr__(self):
        return "Full House"

    def _satisfies(self, dice: list[Die]) -> bool:
        values = [die.value for die in dice]
        return len(set(values)) == 2 and values.count(values[0]) in (2, 3)

    def score(self, dice: list[Die]) -> int:
        return 25 if self._satisfies(dice) else 0

class Straight(_Category):
    SMALL_STRAIGHT = 4
    LARGE_STRAIGHT = 5

    LEGAL_COMBOS = {
        SMALL_STRAIGHT: ({1, 2, 3, 4}, {2, 3, 4, 5}, {3, 4, 5, 6}),
        LARGE_STRAIGHT: ({1, 2, 3, 4, 5}, {2, 3, 4, 5, 6})
    }

    SCORES = {
        SMALL_STRAIGHT: 30,
        LARGE_STRAIGHT: 40
    }

    def __init__(self, length):
        self.length = length

    def __repr__(self):
        return f"{self.length}-Straight"

    def _satisfies(self, dice: list[Die]) -> bool:
        values = {die.value for die in dice}
        for combo in self.LEGAL_COMBOS[self.length]:
            if values.issuperset(combo):
                return True
        return False

    def score(self, dice: list[Die]) -> int:
        return self.SCORES[self.length] if self._satisfies(dice) else 0

class Multiples(_Category):
    def __init__(self, count):
        self.count = count

    def _satisfies(self, dice: list[Die]) -> bool:
        for value in set(die.value for die in dice):
            if sum(1 for die in dice if die.value == value) >= self.count:
                return True
        return False

    def score(self, dice: list[Die]) -> int:
        return sum(die.value for die in dice)

YAHTZEE_CATEGORIES = [
    ONES := UpperCategory(1),
    TWOS := UpperCategory(2),
    THREES := UpperCategory(3),
    FOURS := UpperCategory(4),
    FIVES := UpperCategory(5),
    SIXES := UpperCategory(6),
    CHANCE := Chance(),
    THREE_OF_A_KIND := Multiples(3),
    FOUR_OF_A_KIND := Multiples(4),
    YAHTZEE := Multiples(5),
    FULL_HOUSE := FullHouse(),
    SMALL_STRAIGHT := Straight(4),
    LARGE_STRAIGHT := Straight(5)
]

def demo():
    dice = [Die() for _ in range(5)]
    print("Dice:", dice)
    for die in dice:
        die.roll()
    print("Rolled Dice:", dice)
    print({category : category.score(dice)
           for category in YAHTZEE_CATEGORIES})
demo()
```

```python
import random

class YahtzeeGame:
    def __init__(self, players):
        self.players = players
        self.round_scores = {player: [] for player in players}
        self.scorecards = {player: {category: None for category in ('Ones', 'Twos', 'Threes', 'Fours', 'Fives', 'Sixes', 'Three of a Kind', 'Four of a Kind', 'Full House', 'Small Straight', 'Large Straight', 'Yahtzee', 'Chance')} for player in players}

    def roll_dice(self, num_dice):
        return [random.randint(1, 6) for _ in range(num_dice)]

    def count_dice(self, dice):
        counts = [0] * 7
        for die in dice:
            counts[die] += 1
        return counts[1:]

    def score_yahtzee(self, dice):
        if max(dice) == min(dice):
            return 50
        return 0

    def score_face(self, dice, face):
        return dice.count(face) * face

    def score_category(self, dice, category):
        if category == 'Yahtzee':
            return self.score_yahtzee(dice)
        elif category in ('Ones', 'Twos', 'Threes', 'Fours', 'Fives', 'Sixes'):
            return self.score_face(dice, int(category[-1]))
        elif category == 'Three of a Kind':
            counts = self.count_dice(dice)
            if max(counts) >= 3:
                return sum(dice)
            return 0
        elif category == 'Four of a Kind':
            counts = self.count_dice(dice)
            if max(counts) >= 4:
                return sum(dice)
            return 0
        elif category == 'Full House':
            counts = self.count_dice(dice)
            if 2 in counts and 3 in counts:
                return 25
            return 0
        elif category == 'Small Straight':
            if sorted(set(dice)) in ([1, 2, 3, 4], [2, 3, 4, 5], [3, 4, 5, 6]):
                return 30
            return 0
        elif category == 'Large Straight':
            if sorted(set(dice)) in ([1, 2, 3, 4, 5], [2, 3, 4, 5, 6]):
                return 40
            return 0
        elif category == 'Chance':
            return sum(dice)
        return 0

    def display_scorecard(self, player):
        print("\nCurrent Scorecard for", player)
        for category, score in self.scorecards[player].items():
            if score is None:
                print(f"{category}: Not scored")
            else:
                print(f"{category}: {score}")
        print()

    def reroll_dice(self, dice):
        reroll_indices = list(map(int, input("Enter the indices of dice to reroll (space-separated, starting from 1) or '0' to stop rerolling: ").split()))
        if reroll_indices == [0]:
            return dice
        for index in reroll_indices:
            dice[index - 1] = random.randint(1, 6)
        print("New Dice:", dice)
        return dice

    def play_round(self, round_num):
        round_scores_sum = {}
        for player in self.players:
            print(f"\nRound {round_num} - {player}'s turn:")
            input("Press Enter to roll the dice...")
            dice = self.roll_dice(5)
            print(f"Roll: {dice}")
            for reroll in range(2):
                reroll_choice = input("Do you want to reroll? (yes/no): ").strip().lower()
                if reroll_choice == 'yes':
                    dice = self.reroll_dice(dice)
                    print(f"Reroll {reroll + 1}: {dice}")
                else:
                    break
            self.display_scorecard(player)
            category = input("Enter category to score: ").strip()
            while self.scorecards[player][category] is not None:
                category = input("Category already scored. Enter another category: ").strip()
            score = self.score_category(dice, category)
            self.scorecards[player][category] = score
            self.round_scores[player].append(score)
            self.display_scorecard(player)
            round_scores_sum[player] = sum(self.round_scores[player])
        print("\nRound", round_num, "Scores:", round_scores_sum)

    def play(self):
        for round_num in range(1, 14):
            self.play_round(round_num)
        print("\nFinal Scores:")
        for player in self.players:
            total_score = sum(score for score in self.scorecards[player].values() if score is not None)
            print(f"{player}: {total_score}")

if __name__ == "__main__":
    players = ['Player 1', 'Player 2']
    yahtzee_game = YahtzeeGame(players)
    yahtzee_game.play()
```
